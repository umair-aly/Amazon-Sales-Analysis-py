```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
product_highest_price= data_set.groupby(['ProductName'], as_index=False)['Price'].sum().sort_values(by='Price', ascending=False).head(10)


plt.figure(figsize=(15,7))  ## overall size of the entire chart (canvas)
plt.xlabel('Price',fontweight='bold', fontsize=14)
plt.ylabel('Poduct',fontweight='bold', fontsize=14)
plt.title('Product Price',fontweight='bold', fontsize=24)
plt.xticks(rotation =30)


sns.barplot(
    data=product_highest_price,
    x='ProductName',
    y='Price',
    hue='ProductName',
    palette='cividis',
    legend=False,
    width=0.3   ## It controls the thickness of each bar
    
    
)

plt.show()
```

```python
top_selling_products=data_set.groupby(['ProductName'] , as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False).head(10)
plt.figure(figsize=(15,7))
plt.xlabel('Product Name',fontweight='bold',fontsize=12)
plt.ylabel('Quantity',fontweight='bold',fontsize=12)
plt.xticks(rotation=0)
plt.title("Top 10 Selling Products",fontweight='bold',fontsize=24)

sns.barplot(
    data=top_selling_products,
    y='ProductName',
    x='Quantity',
    width=0.3,
    hue='ProductName',
    palette='cividis',
    legend=False,
)

plt.show()
```

```python
total_revenue=data_set['TotalAmount'].sum()
print('Total Revenue = ',total_revenue)
plt.figure(figsize=(17,7))
plt.xlabel('Total Amount' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('Count', fontweight = 'bold' , fontsize = 14)
plt.title('Revenue Distribution',  fontweight = 'bold' , fontsize = 24)


sns.histplot(
    data_set['TotalAmount'],
    kde=True,
    color = '#001440',
    
   
)
plt.show()
```

```python
revenue_by_product=data_set.groupby(['ProductName'],as_index=False)['TotalAmount'].sum().sort_values(by='TotalAmount',ascending=True)
plt.figure(figsize=(15,7))
plt.xlabel('Prodect Name' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('TotalAmount', fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=90)
plt.title('Revenue by Product', fontweight = 'bold' , fontsize = 24)


sns.lineplot(
    data=revenue_by_product,
    x =  'ProductName',
    y =  'TotalAmount',
    color = '#001440',
    marker='o',
    linewidth=2
)


plt.show()
```

```python

city_sales = data_set.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)

plt.figure(figsize=(17,7))
plt.title("Sales by City",fontweight = 'bold' , fontsize = 24)
plt.xlabel("Revenue", fontweight = 'bold' , fontsize = 14)
plt.ylabel("City", fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)

sns.barplot(
    y=city_sales.index,
    x=city_sales.values, 
    hue = city_sales.values,
    palette='cividis',
    legend=False
)
plt.show()

```

```python

data_set['OrderDate'] = pd.to_datetime(data_set['OrderDate'])
sales_trend = data_set.groupby(data_set['OrderDate'].dt.to_period('M'))['TotalAmount'].sum()


plt.figure(figsize=(15,7))
plt.title("Sales Trend Over Time",fontweight = 'bold' , fontsize = 24)
plt.xlabel("Sales", fontweight = 'bold' , fontsize = 14)
plt.ylabel("Peroid", fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=90)

sns.lineplot(
    x=sales_trend.index.astype(str),  
    y=sales_trend.values,
    marker='o',
    color='#001440',
    linewidth=2
    
)


plt.show()
```

```python
payment_method= data_set['PaymentMethod'].value_counts()

plt.figure(figsize=(15,7))
plt.title("Payment Method Usage",fontweight = 'bold' , fontsize = 24)
plt.xlabel("Count", fontweight = 'bold' , fontsize = 14)
plt.ylabel("Payment Method", fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)


sns.scatterplot(
    y=payment_method.index,
    x=payment_method.values,
    hue=payment_method.values,
    size=payment_method.values,
    legend=False,
    sizes=(100, 5000),
    palette='cividis'
)

plt.show()
```

```python
order_status=data_set['OrderStatus'].value_counts()

plt.figure(figsize=(15,7))
plt.xlabel('Order Status' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('No of Orders' , fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)
plt.title('Order Status Distribution', fontweight = 'bold' , fontsize = 24)

sns.barplot(
    x =  order_status.index,
    y =  order_status.values,
    width = 0.2,
    hue = order_status.values,
    palette ='cividis',
    legend=False,
    
)
plt.show()
```

```python
category_sales=data_set.groupby(['Category'] ,as_index=False)['TotalAmount'].sum().sort_values(by='Category',ascending=False)
plt.figure(figsize=(15,7))
plt.xlabel('Total Amount' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('Category' , fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)
plt.title('Category Performance', fontweight = 'bold' , fontsize = 24)


sns.barplot(
    data = data_set,
    x = 'TotalAmount',
    y =  'Category',
    width = 0.3,
    hue = 'Category',
    palette ='cividis',
    legend=False,
    
)
plt.show()

```

```python
discount_effect = data_set.groupby(['Discount'], as_index=False)['TotalAmount'].mean()
plt.figure(figsize=(15,7))
plt.xlabel('Discount' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('Total Amount' , fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)
plt.title('Discount Impact', fontweight = 'bold' , fontsize = 24)

sns.lineplot(
    data=discount_effect,
    x='Discount',
    y='TotalAmount', 
    marker='o',
    color='#001440'
)

plt.show()
```

```python
discount_effect = data_set.groupby(['Brand'], as_index=False)['TotalAmount'].mean()
plt.figure(figsize=(15,7))
plt.xlabel('Total Amount' , fontweight = 'bold' , fontsize = 14)
plt.ylabel('Brand' , fontweight = 'bold' , fontsize = 14)
plt.xticks(rotation=0)
plt.title('Revenue by Brand', fontweight = 'bold' , fontsize = 24)

sns.barplot(
    data=discount_effect,
    x='TotalAmount',
    y='Brand',
    hue= 'Brand',
     width = 0.4,
    palette ='cividis',
    legend=False, 
    linewidth=2
    
)

plt.show()
```
